```python
import numpy
import csv
with open('rsel.csv') as csvfile:
    R=csv.reader(csvfile,delimiter=',')
    i=0
    rsel=[]
    rselr=[]
    for row in R:
        if i>0:
            f=[]
            for j in range(len(row[2:])):
                f.append(float(row[j+2]))
            rsel.append(numpy.mean(f))
            rselr.append(float(row[1]))
        i=i+1
```

```python

with open('cel.csv') as csvfile:
    R=csv.reader(csvfile,delimiter=',')
    i=0
    cel=[]
    celr=[]
    for row in R:
        if i>0:
            f=[]
            for j in range(len(row[2:])):
                f.append(float(row[j+2]))
            cel.append(numpy.mean(f))
            celr.append(float(row[1]))
        i=i+1
```

```python
with open('cel-rs.csv') as csvfile:
    R=csv.reader(csvfile,delimiter=',')
    i=0
    cels=[]
    celsr=[]
    for row in R:
        if i>0:
            f=[]
            for j in range(len(row[2:])):
                f.append(float(row[j+2]))
            cels.append(numpy.mean(f))
            celsr.append(float(row[1]))
        i=i+1
```

```python
with open('2cel-rs.csv') as csvfile:
    R=csv.reader(csvfile,delimiter=',')
    i=0
    zcels=[]
    zcelsr=[]
    for row in R:
        if i>0:
            f=[]
            for j in range(len(row[2:])):
                f.append(float(row[j+2]))
            zcels.append(numpy.mean(f))
            zcelsr.append(float(row[1]))
        i=i+1
```

```python
with open('2cel.csv') as csvfile:
    R=csv.reader(csvfile,delimiter=',')
    i=0
    zcel=[]
    zcelr=[]
    for row in R:
        if i>0:
            f=[]
            for j in range(len(row[2:])):
                f.append(float(row[j+2]))
            zcel.append(numpy.mean(f))
            zcelr.append(float(row[1]))
        i=i+1
```

```python
import matplotlib.pyplot as plt

def main():
    plt.figure(figsize=(6.7,6.7))
    
    plt.plot(rselr,rsel,'b',label='1-Evol-RS')
    plt.plot(celsr,cels,'g',label='1-Coev-RS')
    plt.plot(zcelsr,zcels,'r',label='2-Coev-RS')
    plt.plot(celr,cel,'k',label='1-Coev')
    plt.plot(zcelr,zcel,'m',label='2-Coev')
    
    plt.axis([0, 500000,0.60, 1])
    plt.xlabel('Rozegranych gier')
    plt.ylabel('Odsetek wygranych gier [%]')
    
    plt.legend(loc='lower right',fontsize='xx-large')
    
    plt.savefig('myplot.pdf')
    plt.close()
    
if __name__ == '__main__':
    main()
```
