```python
import numpy as np

from sympy.interactive import printing
printing.init_printing(use_latex=True)

from __future__ import division
import sympy as sym
from sympy import *

x,y=symbols('x y')
```

```python
eq=Eq(-x**3+3*x**2+10*x-24)
solve(eq,x)
```

```python
from matplotlib import pyplot as plt

z=np.linspace(-5,5,101)
es=(-z**3+3*z**2+10*z-24)
plt.plot(z,es)
plt.show()
```

```python
ex=Eq(x**2+3*y,10)
ey=Eq(4*x-y**2,-2)
solve((x**2+3*y-10,4*x-y**2+2), x, y)
```
